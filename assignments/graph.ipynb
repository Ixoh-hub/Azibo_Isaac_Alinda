```python
import pandas as pd
import matplotlib.pyplot as pylot
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
pylot.figure(figsize=(8,5))
sns.histplot(df['Maxpulse'], kde=True, color ='skyblue')
pylot.title('Maxpulse Distribution')
pylot.xlabel('Maxpulse')
pylot.ylabel('Count')
pylot.show()
```

```python
pylot.figure(figsize=(8, 5))
sns.histplot(df['Calories'], kde=True, color='lightgreen')
pylot.title('Calories Distribution')
pylot.xlabel('Calories')
pylot.ylabel('Count')
pylot.show()
```

```python
# side by side view

pylot.figure(figsize=(14, 5))

# Maxpulse Histogram
pylot.subplot(1, 2, 1)
sns.histplot(df['Maxpulse'], kde=True, color='skyblue')
pylot.title('Maxpulse Distribution')
pylot.xlabel('Maxpulse')
pylot.ylabel('Count')

# Calories Histogram
pylot.subplot(1, 2, 2)
sns.histplot(df['Calories'], kde=True, color='lightgreen')
pylot.title('Calories Distribution')
pylot.xlabel('Calories')
pylot.ylabel('Count')

pylot.tight_layout()
pylot.show()
```
